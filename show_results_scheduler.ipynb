```python
from pathlib import Path
from typing import Any, Tuple

import torch

from matplotlib import pyplot as plt

from pinn.run_training import get_config_hash

from pinn.problems_definitions import (BuckleyLeverett, NonLinearNonConvexFlow,
                                       PeriodicSine2D, Problem, Rarefaction1D,
                                       Riemann2D, RiemannOblique, Shock1D, Pulse)
from pinn.slope_limiters import (advection_residual_autograd,
                                 advection_residual_mm2,
                                 advection_residual_mm3,
                                 advection_residual_uno)
from pinn.training import Config


def load_results(
    problem: Problem, config: Config, results_dir: Path = Path("results")
) -> Tuple[Problem, Any]:
    hash_id = get_config_hash(problem, config)
    model_path = results_dir / f"model_{hash_id}.pth"
    plot_path = results_dir / f"plot_{hash_id}.png"

    model_state = torch.load(model_path)
    problem.net.load_state_dict(model_state)

    training_history = plt.imread(plot_path)
    return problem, training_history

from IPython.display import HTML, Video
from typing import List

def display_videos_with_images(pairs: List[Tuple[str, str]]) -> HTML:
    rows = []
    for i in range(0, len(pairs), 4):
        group = pairs[i:i+4]
        row_html = "<div style='display: flex; justify-content: space-around; margin-bottom: 20px;'>"
        for video_path, image_path in group:
            row_html += f"""
            <div style="text-align: center; margin: 5px;">
                {Video(video_path, embed=True)._repr_html_()}
                <img src="{image_path}" style="max-width: 100%; margin-top: 5px;">
            </div>
            """
        row_html += "</div>"
        rows.append(row_html)
    return HTML("".join(rows))

problem = Pulse
```

```python
configs = [Config(
    epsilon=0.0025,
    n_points=128000,
    epochs=15000,
    residual=res,
) for res in [
    advection_residual_autograd,
    advection_residual_mm2,
    advection_residual_mm3,
    advection_residual_uno
    ]
]

hash_ids = [get_config_hash(problem(), config) for config in configs]

video_image_pair = [
    (
        f"videos/animate_problem_{hash_id}.mp4",
        f"results/plot_{hash_id}.png",
    )
    for hash_id in hash_ids
]

display_videos_with_images(video_image_pair)
```
